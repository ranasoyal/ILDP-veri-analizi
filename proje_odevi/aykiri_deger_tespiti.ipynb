```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV dosyasını yükle
dataset = pd.read_csv("Indian Liver Patient Dataset (ILPD).csv")

# Box plot oluştur
plt.figure(figsize=(10, 6))
dataset.boxplot()
plt.title('Box Plot')
plt.xticks(rotation=45)
plt.show()
```

```python
# sgot değeri 2946 olan gözlemleri filtrele ve sil
dataset_yeni = dataset[dataset['sgot'] != 2946]

# Aykırı değerlerden temizlenmiş veri setini kaydetme
dataset_yeni.to_csv("Indian Liver Patient Dataset (ILPD)-2.csv", index=False)
```
