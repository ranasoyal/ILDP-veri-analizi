```python
import pandas as pd
import seaborn as sns # Veri grselletirme iin kullanlan bir ktphane.
import matplotlib.pyplot as plt # Temel grafiklerin izilmesi iin kullanlr.

# CSV dosyasn ykle
dataset = pd.read_csv("Indian Liver Patient Dataset (ILPD).csv")

# Korelasyon matrisini oluturulur
korelasyon_matrisi = dataset.corr()

# Heatmap oluturulur
plt.figure(figsize=(10, 8))
sns.heatmap(korelasyon_matrisi, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Korelasyon Matrisi')

# Korelasyon katsaylar her kare iine yazdrlr
for i in range(len(korelasyon_matrisi)):
    for j in range(len(korelasyon_matrisi)):
        plt.text(j + 0.5, i + 0.5, round(korelasyon_matrisi.iloc[i, j], 2),
                 color="black", ha='center', va='center', fontsize=10)

plt.show()

# Korelasyon matrisini yazdr
# print(korelasyon_matrisi)
```
